```python
seq = input("Masukkan sekuen dna :")
n_A = seq.count("A")
n_T = seq.count("T")
n_G = seq.count("G")
n_C = seq.count("C")
k_GC = n_G + n_C/len(seq)*100
print ("n_A=", n_A, "n_T=", n_T, "n_G=", n_G, "n_C=", n_C, "k_GC=", k_GC)
```
